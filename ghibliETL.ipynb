```python
import pandas as pd
import requests
```

# Extração

```python
url = "https://ghibliapi.vercel.app/films"
res = requests.get(url)
data = res.json()
```

```python
df = pd.DataFrame(data)

print("Dados:")
display(df.head())
```

# Transformação

```python
df = df[["title", "original_title", "director", "release_date", "running_time", "rt_score"]]

df["release_date"] = pd.to_numeric(df["release_date"], errors="coerce")
df["running_time"] = pd.to_numeric(df["running_time"], errors="coerce")
df["rt_score"] = pd.to_numeric(df["rt_score"], errors="coerce")

df = df.sort_values(by="rt_score", ascending=False)

print("Dados transformados:")
display(df)
```

# Carga

```python
df.to_csv("ghibli_filmes_transformado.csv", index=False)

print("Arquivo salvo: ghibli_filmes_transformado.csv")
```
